```python
import re

def clean_text(string_in):
    string_in = re.sub(r"[^A-Za-z]+", "", string_in) # Remove special characters
    string_in = re.sub(r"\d", "", string_in)   # Remove numbers 
    string_in = re.sub(r"spam", "", string_in)   # Remove "spam" word
    string_in = string_in.lower()              # Tranform to lower case    
    
    return string_in.strip()

import nltk
from nltk import word_tokenize
f = open("SMSSpamCollection.txt", "r")

spamArr = []

for x in f:
    tokens = word_tokenize(x)
    if tokens[0] == "spam":
        for token in tokens:
            text = clean_text(token)
            if text != '':
                spamArr.append(clean_text(token))
        
print(spamArr)  
```

```python
import pandas as pd 

# Python code to find frequency of each word 
def freq(str): 
              
    for i in range(0, len(spamArr)): 
  
        # count the frequency of each word present in spamArr and print 
        print(spamArr[i], ':', spamArr.count(spamArr[i]))
        
freq(spamArr)        
```
